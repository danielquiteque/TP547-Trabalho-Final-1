#  Visualização de Resultados – IA para Energia em Edifícios
Este notebook demonstra os gráficos principais usados para analisar o desempenho do sistema baseado em IA no gerenciamento de energia.

```python
import matplotlib.pyplot as plt
import numpy as np
```

##  Gráfico 1: Previsão de Consumo com LSTM

```python
# Simulação de dados
y = np.sin(np.linspace(0, 10, 100)) + 0.1 * np.random.randn(100)  # real
preds = y + 0.05 * np.random.randn(100)  # previsão

plt.figure(figsize=(12, 5))
plt.plot(y, label="Real", linewidth=2)
plt.plot(preds, label="Previsto", linestyle="--")
plt.xlabel("Tempo")
plt.ylabel("Consumo de Energia (normalizado)")
plt.title("Consumo de Energia - Real vs Previsto (LSTM)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Gráfico 2: Redução de Consumo por Categoria

```python
labels = ["HVAC", "Iluminação", "Total"]
reducoes = [29.7, 23.4, 33.8]  # porcentagens

plt.figure(figsize=(8, 5))
plt.bar(labels, reducoes, color=["#4CAF50", "#2196F3", "#FFC107"])
plt.title("Redução no Consumo de Energia (%)")
plt.ylabel("Redução (%)")
plt.ylim(0, 40)
for i, v in enumerate(reducoes):
    plt.text(i, v + 1, f"{v:.1f}%", ha="center", fontweight="bold")
plt.tight_layout()
plt.show()
```

##  Gráfico 3: Consumo Diário Antes vs. Depois da Otimização

```python
dias = list(range(1, 31))
consumo_antes = np.random.normal(loc=200, scale=10, size=30)
consumo_depois = consumo_antes * 0.75  # 25% de economia

plt.figure(figsize=(12, 5))
plt.plot(dias, consumo_antes, label="Antes (kWh)", color="red")
plt.plot(dias, consumo_depois, label="Depois (kWh)", color="green")
plt.title("Consumo Diário de Energia: Antes vs Depois da Otimização com IA")
plt.xlabel("Dias")
plt.ylabel("Consumo de Energia (kWh)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
